TO FILL IN THE MISSING VALUES

```python
#MEAN
mean_value_bdr=d['No. Bedrooms'].mean()
d['No. Bedrooms'].fillna(value=mean_value_bdr, inplace=True)
mean_value_btr=d['No. Bathroom'].mean()
d['No. Bathroom'].fillna(value=mean_value_btr, inplace=True)
mean_value_sqft=d['sq.ft'].mean()
d['sq.ft'].fillna(value=mean_value_sqft, inplace=True)
print('NaN values replaced by Mean values')
print(d)
```

```python
#MEDIAN
df=pd.read_excel('ramshree.xlsx')
df['No. Bedrooms'] = df['No. Bedrooms'].fillna(df['No. Bedrooms'].median())
df['No. Bathroom'] = df['No. Bathroom'].fillna(df['No. Bathroom'].median())
df['sq.ft'] = df['sq.ft'].fillna(df['sq.ft'].median())
df
```

```python
#MODE
df=pd.read_excel('ramshree.xlsx')
df['No. Bedrooms'] = df['No. Bedrooms'].fillna(df['No. Bedrooms'].mode()[0])
df['No. Bathroom'] = df['No. Bathroom'].fillna(df['No. Bathroom'].mode()[0])
df['sq.ft'] = df['sq.ft'].fillna(df['sq.ft'].mode()[0])
df
```

```python
## ENCODING means creating 'duplicate' 
## encoding also means converting data into a different form 
## There are three types of encoding 
## 1. one-hot encoding-  In one hot encoding, for every categorical feature, a new variable is created
## 2. ordinal encoding - converts each label into integer values and the encoded data represents the sequence of labels
## 3. label encoding- encodes the levels of categorical features into numeric values.
```
